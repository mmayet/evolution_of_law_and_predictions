```python
class Books():
    def __iter__(self):
        for i,file_name in enumerate(glob('../../data/134/*.json')):
            try:
                with open(str(file_name)) as f:
                    book_text = json.load(f)['text']
                
                #### Start Processing
                st = ISRIStemmer()
                stemmed_book = st.stem(book_text)
                print(str(file_name))
                yield TaggedDocument(stemmed_book,[i])

            except:
                print("Fix {}".format(file_name))
```

```python
def not_small(text):
    return len(text) >= 2

def not_big(text):
    return len(text) < 10

def not_sw(text):
    arb_stopwords = stopwords.words("arabic")
    letters = ['ا','أ','إ','ذ','ض','ص','ث','ق','ف','غ','ع','ه','خ','ح','ج','ش','س','ي','ب','ل','ا','ال','ت','ن','م','ك','ئ','ء','ؤ','ر','لا','ى','ة','و','ز','ظ']
    words = ['قال','عليه','الله','قوله','له','قوله','ولو','صلى','وسلم','لأنه','فلا','ثم','به','كان','','','','','','','','','','','','','',]
    sw = set(arb_stopwords + letters + words)
    return text not in sw
```

```python
texts = []

for i,file_name in enumerate(glob('../../data/134/*.json')):
    with open(str(file_name)) as f:
        raw = json.load(f)['text']
    
    texts.append(araby.tokenize(raw, conditions=[not_sw,not_small,not_big,araby.is_arabicword,araby.is_arabicstring], morphs=[araby.strip_tashkeel,araby.strip_shadda,araby.strip_tatweel]))

# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)
    
# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
lda = LdaModel(corpus, num_topics=5, 
                            id2word=dictionary, 
                            update_every=5, 
                            chunksize=100, 
                            passes=10)
```

```python
print(lda.print_topics(num_topics=5, num_words=25))
```
